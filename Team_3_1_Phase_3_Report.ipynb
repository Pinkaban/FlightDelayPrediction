## Abstract

This project addresses the business problem of predicting flight departure delays of 15 minutes or more, a critical challenge for airlines, airports, and passengers seeking to improve operational efficiency and reduce uncertainty. The analysis leverages the OTPW (On-Time Performance with Weather) dataset, which integrates U.S. Department of Transportation (DOT) airline on-time performance data with NOAA weather observations. Two dataset scopes were explored: a 3-month subset for rapid exploration and a 1-year dataset to capture seasonal patterns and improve generalizability.

As a baseline, a Logistic Regression model was selected due to its interpretability, efficiency, and strong performance on binary classification tasks. This model provides a transparent benchmark against which more complex models can be evaluated. Across the project, six experiments were conducted to iteratively improve model performance. The best-performing experiment incorporated additional feature engineering, leading to a peak PR-AUC of 0.3595. Performance was evaluated using metrics aligned with the problem’s priorities, particularly the F2-score (to emphasize recall of delayed flights), AUC, and PR-AUC. Feature importance analysis highlighted variables such as departure hour, route-level historical delay rates, and weather-related factors (e.g., visibility, precipitation, and wind conditions) as the most influential predictors.

Looking ahead to Phase III, several enhancements are planned to improve predictive performance. These include implementing more advanced models such as gradient boosting and random forests, expanding feature engineering (e.g., lagged delays, airport congestion metrics, and interaction terms), and fully leveraging the 1-year dataset for training. Another advanced model that we are planning to use is the Multi-Layer Perceptron Model (MLP) to enhance results. Additional efforts will focus on systematic hyperparameter tuning and model calibration to optimize performance under the F2 objective. Together, these improvements aim to deliver a more robust and operationally useful delay prediction system.



## Project Information

### Project Description
Flight delays driven by technical aircraft issues, operational bottlenecks, and adverse weather represent a massive logistical challenge. The aviation industry faces over $30 billion in annual costs from flight delays in the U.S. alone, with the global economic impact of disruptions reaching up to $67.5 billion when factoring in lost passenger time and operational overhead. These expenses include wasted fuel, crew time, maintenance, and lost productivity (Anupkumar, 2023).

Our analysis draws on multiple complementary datasets spanning the full calendar year of 2015. The primary dataset is the OTPW (On Time Performance and Weather) table, which is a pre joined source of approximately 5.8 million rows and 216 columns combining U.S. DOT Bureau of Transportation Statistics flight records with NOAA hourly weather observations. This is supplemented by the individual Flights dataset containing carrier, route, schedule, and delay information. We also utilize the Weather dataset for hourly meteorological readings including wind speed, visibility, precipitation, temperature, and sky conditions, as well as the Stations dataset for weather station metadata and airport to station mappings, alongside Airport and Airline reference data.

The target variable is DEP_DEL15, a binary indicator that equals 1 when a flight departure delay exceeds 15 minutes. The dataset exhibits a natural class imbalance of approximately 80 percent on time flights to 20 percent delayed flights. Through feature engineering, we expanded the raw inputs into 41 model ready features organized into three families. These include temporal features like the quarter, month, day of week, and departure hour. We also included meteorological features such as wind speed, visibility, precipitation, and derived indicators for thunderstorms, rain, snow, fog, freezing precipitation, and haze. Finally, we engineered operational features including departure sequencing, cumulative same day delay rates, multi horizon delay propagation windows, holiday proximity buckets, and one hot encoded categorical variables.

We frame this as a binary classification problem, predicting whether a given flight will be delayed by 15 minutes or more at departure. Our methodology begins with extensive Exploratory Data Analysis of the OTPW, Flights, Weather, and Stations datasets, examining summary statistics, missing data patterns, cardinality, redundancy, and outliers. These EDA insights inform targeted feature engineering across both flight operations and weather dimensions.

For modeling, we employ a distributed Logistic Regression pipeline in PySpark with L2 regularization as our baseline classifier. To address class imbalance, we apply dynamic random undersampling strictly to training folds, forcing a 1:1 balance. The evaluation framework uses strict chronological temporal splits, utilizing January through September for training, October for validation, and November through December for testing to prevent data leakage. We pair this with Sliding Window Time Series Cross Validation to mitigate concept drift. The decision threshold is tuned on the validation set to maximize the F2 Score, which weights recall twice as heavily as precision, perfectly reflecting the high cost of missing actual delays in aviation operations. Looking ahead, we are building a custom flight and weather join with station aligned, time synchronized weather and tail level turnaround metrics to serve as a richer feature foundation for more advanced models in Phase III.

* *A cancelled flight never departs, so all post-departure fields are physically impossible
to populate.
* *Flight Was On Time — No Delay to Attribute (~81% of delay-cause fields)
The five delay-cause columns are only populated when BTS determines a **reportable delay
occurred**. For the ~12M on-time or early flights, there is simply no delay to categorize.
* *Multiple Diversions Are Extremely Rare (~100% for DIV2–DIV5)


**Missing Data Strategy:**
* For columns :**`CARRIER_DELAY`, `WEATHER_DELAY`, `NAS_DELAY`, `SECURITY_DELAY`,
  `LATE_AIRCRAFT_DELAY` → **~12M nulls (81.28%)**

> These nulls mean "zero minutes of delay", not missing data. We are excluding these columns from the modeling dataset because they are known after the delay happens.

#### Cardinality Analysis


Most variables (e.g., DAY_OF_WEEK, MONTH) have low distinct counts, which is ideal for modeling since they can be directly used as categorical predictors without transformation. These variables capture structural patterns in delays (seasonality, operational disruptions, etc.) and are strong candidates to keep. There are other columns, such as cancellation-related or diversion-related variables, that also show low cardinality because they are populated for very few observations. FLIGHTS is the only variable that has no missing values, but its values are constant, so this variable does not add value for modeling.

On the other hand, TAIL_NUM and OP_CARRIER_FL_NUM stand out as the only variables with higher cardinality compared to the rest of the dataset, although still low in relative terms (< 1% uniqueness). TAIL_NUM contains meaningful signal because it reflects aircraft-specific factors such as reliability, maintenance, and operational patterns, which can influence delay behavior. However, using it directly may increase the risk of overfitting and poor generalization.

#### Outlier Analysis

The outlier analysis shows that most temporal variables (e.g., QUARTER, MONTH, DAY_OF_WEEK, and scheduled times) do not exhibit any outliers. This is expected, as these variables are either bounded by definition or represent discrete identifiers rather than continuous measurements. Similarly, several operational indicators such as CANCELLED, DIVERTED, and FLIGHTS also show no outliers, reflecting their limited variability or binary/constant nature.

In contrast, delay-related and operational performance variables exhibit a significant presence of outliers. Variables such as `DEP_DELAY_NEW` (16.9%), `ARR_DELAY_NEW` (17.3%) show the highest proportion of outliers, which is consistent with the inherently skewed nature of delays, where extreme values (long delays) are relatively frequent. Other operational metrics such as `TAXI_OUT`, `TAXI_IN`, `AIR_TIME`, `DISTANCE`, and elapsed times also present moderate levels of outliers (around 5–6%), likely reflecting variability in airport congestion, route characteristics, and operational disruptions. Additionally, delay components like `CARRIER_DELAY`, `NAS_DELAY`, and `LATE_AIRCRAFT_DELAY` show smaller but still meaningful outlier presence, indicating occasional extreme disruptions in specific delay categories.

Overall, the presence of outliers is concentrated in variables that capture operational variability and disruption intensity, which are precisely the signals most relevant for predicting delays. Rather than being treated as noise, these outliers likely represent meaningful extreme events (like severe congestion, weather impacts, or cascading delays).


#### Overall Learnings

This exploratory data analysis helped us understand the key drivers of flight delays and identify patterns that can inform the development of the model. This includes analyzing temporal trends (e.g., seasonality, time of day), operational factors (e.g., airlines, airports, routes), and delay causes to determine which variables have the strongest relationship with delay occurrence. Additionally, the analysis uncovered how delays propagate over time—both within a single day and across multiple days—by exploring dynamic features such as morning performance and prior-day conditions.

The following insights validate why some variables should be included in the model:

* The main cause of delays is carrier-related issues, accounting for nearly half of all delayed flights. While weather represents only about 3% of delays, it may still be an important predictor, particularly for identifying delays driven by extreme conditions.

![route_delays](https://raw.githubusercontent.com/Pinkaban/FlightDelayPrediction/main/Phase%20II%20Report%20Images/main_delay_reason.png)

* Quarter and month appear to be strong predictors. The average quarterly delay rate is 18.6%, with the second quarter standing out, showing an 11% higher delay rate than the others. This may be related to increased travel demand during summer vacation periods. At a monthly level, June exhibits a delay rate 28% above the average. February and December also show above-average delays, likely influenced by harsher weather conditions. In contrast, September, October, and November have the lowest delay rates.

* Since carrier delays are the dominant cause, we analyzed delay rates by airline. While the average delay rate is 20.5%, there is significant variation across carriers. For example, Frontier Airlines (F9) has the highest delay rate, approximately 51% above the average, while Hawaiian Airlines shows the lowest rate, with only 8.8% of flights delayed.

* Looking at airport-level patterns, we observe clear bottlenecks. While the average airport delay rate is around 20%, Chicago O'Hare International Airport (ORD) reaches 30.4% (52% above average). In contrast, Daniel K. Inouye International Airport (HNL) has a delay rate of only 10% (53% below average). This suggests that airport-specific factors are important predictors of delay. This lead us to the hypothesis that hub airports (e.g., ATL, DFW, DEN, ORD, LAX) would exhibit higher delay rates due to their high traffic volume. However, when comparing low vs high volume airports, the difference in delay rates is only about 10%, suggesting that factors beyond volume play a more important role.

* At the route level, we expected more popular origin-destination pairs to have higher delay rates. Instead, we found that low-volume routes exhibit high variability, while high-volume routes converge toward the average delay rate (~20%), as seen in popular routes like LAX–SFO, LAX–JFK, and LGA–ORD.

![route_delays](https://raw.githubusercontent.com/Pinkaban/FlightDelayPrediction/main/Phase%20II%20Report%20Images/flights%20volume.png)

* Finally, “domino-effect” variables are likely to improve delay prediction because they capture how disruptions propagate over time. Intra-day features reflect how early-morning performance impacts delays later in the same day, showing clear accumulation effects. Inter-day features capture persistence from the previous day, indicating that disruptions carry over due to operational constraints such as aircraft rotations. The X-day lag propagation variables extends this idea by measuring recent system stress, which consistently correlates with higher delay rates. Together, these features provide dynamic, time-dependent context that enhances predictive performance beyond static variables.

**Handling Strategy:** We ultimately dropped 99 columns that were missing over 94% of their data, which were mostly daily and monthly aggregates. We chose to keep the event driven fields because the nulls serve as actually meaningful signals. For our modeling purposes, null precipitation and gusts will be imputed as zero, while null temperatures and pressures will be forward filled or replaced with the specific station level mean.



![Selected Columns Missing Data](https://raw.githubusercontent.com/Pinkaban/FlightDelayPrediction/main/Phase%20II%20Report%20Images/weather_selected_columns_missing.png)

The missing data here is not random, but rather follows a clear tiered pattern based on how stations report and what sensors they actually have installed.
* Critical Missingness (over 80% missing): This tier includes features like HourlyWindGustSpeed, HourlyPresentWeatherType, and HourlyPrecipitation. These measurements are entirely event driven. For instance, gust speed is only recorded when it actually gusts, and precipitation is only logged when it rains or snows. In these specific cases, a null value essentially just means that no weather event occurred.
* High Missingness (40 to 72% missing): This category includes sea level pressure, sky conditions, wet bulb temperature, and station pressure. The high missingness occurs because many of the smaller weather stations simply lack the physical sensors required to record these metrics.
* Moderate Missingness (25 to 38% missing): This group includes standard aviation METAR fields like visibility, altimeter setting, humidity, and sustained wind speed. While they are standard measurements, they are not universally reported every single hour across all weather stations.
* Low Missingness (under 5% missing): This tier includes basic identifiers, geographical coordinates, and the ambient temperature (HourlyDryBulbTemperature).

#### Cardinality Analysis
In this section, we assess the cardinality (number of unique values) and the internal distributions of our categorical features. Because the global airport dataset is highly diverse, understanding these distributions is critical to filtering down the dataset to only the airports relevant to our domestic US flight delay model.

**Key Findings:**
* Airport Type Distribution (Low Cardinality, High Skew): The `type` column contains only 7 distinct categories. However, the dataset is heavily dominated by `small_airport` (60.62%) and `heliport` (20.95%). Commercial airline operations, which our flight dataset tracks, almost exclusively occur at `medium_airport` (7.9%) and `large_airport` (1.07%). 
* Country Skewness (High Cardinality, High Skew): The `iso_country` column possesses high cardinality, representing hundreds of distinct global regions. Despite this spread, the distribution is extremely skewed: the US accounts for a massive 40.5% (23,256 airports) of the global total, followed distantly by Brazil (8.78%) and Canada (4.87%).
* Primary Key Cardinality Constraint Violation: The `iata_code` column is intended to act as our unique primary key to join against the flights dataset. However, our duplicate analysis revealed 39 specific `iata_code` values that map to multiple rows (e.g., the placeholder code `0` maps to 75 rows, and `AHT` maps to 2).


#### Redundancy Analysis

In this section, we analyzed the dataset for redundant features to optimize our PySpark joins and reduce memory bloat. Specifically, we investigated the identifier columns (`ident`, `iata_code`, and `local_code`) and regional columns (`iso_region` vs. `iso_country`).

**Key Findings:**
* Identifier Overlap: The `ident` column perfectly matches the `iata_code` 594 times, and the `local_code` matches it 1,788 times. Since `iata_code` is the required primary key for joining with our Phase 1 Flights dataset, the other identifiers are either strictly redundant (when they match) or irrelevant to our join condition (when they differ). 
* Regional Redundancy: The `iso_region` column contains the `iso_country` prefix exactly **100.00%** of the time (e.g., a US airport in California is labeled `US-CA`, duplicating the `US` country code). 



#### Outlier Analysis

Aviation datasets often contain "natural" outliers. An airport elevation of 6,000 feet is an extreme statistical outlier compared to sea-level airports, but it represents a real geographic feature, not an error. 

Below, we performed an Interquartile Range (IQR) analysis on the continuous numerical variable `elevation_ft` to identify statistical outliers, followed by logical bound checks to identify potential data-entry errors.

**Key Findings:**
* Statistical Outliers: The IQR analysis identified 4,747 airports (9.57% of the dataset) as statistical outliers, falling above the upper bound of 3,387.5 ft. In aviation, this is expected. High-altitude airports (like Denver International at 5,431 ft or airports in the Andes) legitimately exist outside the standard distribution. 
* Logical Extremes: Our sanity check (filtering for elevations > 15,000 ft and < -1,500 ft) caught three extreme anomalies:
  * *Siachen Glacier AFS* (22,000 ft) & *Daulat Beg Oldi* (16,200 ft): These are genuine high-altitude military bases in the Himalayas.
  * *"Modi" in the US* (29,977 ft): This is a clear data-entry error. The highest peak in the United States is Denali (~20,310 ft), making a 29,977 ft airport physically impossible (it is nearly the cruising altitude of a commercial jet).


#### Overall Learnings

Based on the Exploratory Data Analysis performed on the Global Airports dataset, we have formulated the following actionable, strict rules for the Phase 2 Feature Engineering pipeline. These rules ensure cluster memory optimization and prevent data corruption when joining with the primary Flights dataset.

1. Scope Reduction (Filtering by Cardinality):
The global dataset contains massive geographical and operational noise. Because our predictive model focuses on US commercial aviation, we must aggressively filter the dataset prior to joining:
* Country Filter: Restrict the dataset strictly to `iso_country == 'US'`.

2. The Join Key Mandate (Preventing Cartesian Explosions):
The `iata_code` is our sole primary key for joining with the BTS flights data. However, it violates unique constraints.
* Rule: Drop all rows where `iata_code` is null. 
* Rule: Explicitly de-duplicate the dataset based on `iata_code` to resolve the 39 overlapping keys (e.g., the placeholder code `0`) *before* executing a PySpark `.join()`. Failing to do so will result in a Cartesian explosion of false rows.

3. Feature Pruning (Handling Redundancy & Sparsity):
To optimize distributed memory limits, we will drop columns that provide no unique signal:
* Redundant IDs: Drop `ident` and `local_code` as they either duplicate `iata_code` or are useless for the join.
* Geographic Redundancy: Drop `iso_region` as it redundantly contains the `iso_country` prefix.


**4. Outlier Strategy (Domain-Specific Sanity):**
* Rule: Do NOT drop statistical outliers in the `elevation_ft` column. The ~9.5% of airports falling above the IQR upper bound represent legitimate geographical topography (e.g., Denver International) which impacts flight physics and delays. 
* Rule: Apply a logical cap to drop physically impossible data-entry errors (e.g., dropping the "Modi" US airport erroneously logged at 29,977 ft).

#### Missing Data Analysis

An assessment of missingness across the OTPW dataset reveals substantial variability in data completeness across feature groups. A large subset of fields—primarily monthly climate summaries, short-duration precipitation metrics, and derived climatological normals—exhibit 100% missingness, indicating that these fields were likely not captured, and not available for our modeling purpose.

Operational and flight-level fields show significantly higher completeness, with most aviation-critical variables nearly fully populated. Several backup equipment configuration fields, present/weather-type indicators, and hourly precipitation fields exhibit moderate to high missingness due to inconsistent reporting across stations. Overall, the data quality is strong for core operational, temporal, and engineered features, while purely climatological and infrequently reported station-level attributes contribute disproportionately to the missing data.

The table below provides a summary view of the missingness in the dataset

| Missingness Range | Number of Columns | Examples                                                                         |
| ----------------- | ----------------- | -------------------------------------------------------------------------------- |
| **100% missing**  | 60+               | MonthlyMeanTemperature, ShortDurationPrecipitationValue060, MonthlyTotalSnowfall |
| **90–100%**       | ~4                | DailyWeather, DailyPrecipitation, Cancellation Code                              |
| **60–90%**        | ~5                | BackupEquipment, BackupElements                                                  |
| **10–60%**        | 1–2               | HourlyPrecipitation                                                              |
| **0–10%**         | Most columns      | TAIL_NUM, REM, operational fields                                                |

The missingness profile shows a clear two-tier data landscape:

Tier 1: High-Value, High-Completeness Data (~0-10% missing)

This includes:
- Flight metadata
- Temporal features
- Engineered weather features
- Hourly weather observations
- Network and congestion metrics
- Tail and aircraft information

These form the core of most modeling and analysis tasks and display excellent data availability.

Tier 2: Climatological & Rarely Reported Fields (≥90% missing)

This includes:
- All monthly climate statistics
- Short-duration precipitation metrics
- Climatology normals
- Daily weather summaries

These fields are effectively unusable in their current form and should be deprioritized or removed.

#### Cardinality Analysis

To understand the uniqueness and variability of each column, a cardinality ratio was calculated for all columns. 

Columns with high cardinality ratios (close to 1) are highly unique, while low ratios indicate repeated or categorical values. This analysis helps identify key identifiers, categorical features, and potentially low-variance columns that may need special treatment in modeling or reporting.

| Metric                               | Value / Range                                               |
| ------------------------------------ | ----------------------------------------------------------- |
| Columns with cardinality ratio >0.05 | 6-10 columns (unique per flight/timestamp)                  |
| Columns with 0.01–0.05               | ~20-30 columns (moderate uniqueness, weather, date)         |
| Columns with <0.01                   | Majority (~180 columns, low variability, categorical codes) |
| Max cardinality ratio                | 0.1958 (REM)                                                |
| Min cardinality ratio (non-zero)     | 0.000000344 (FLIGHTS, YEAR)                                 |
| Number of identifier columns         | ~5–10                                                       |

The cardinality analysis provides following insight:
1. Identifier columns such as `FL_DATE`, `OP_CARRIER_FL_NUM`, `TAIL_NUM`, and `STATION` are effectively unique per flight or station but show low ratio due to large dataset size.
2. Most operational, airport, and flight status fields (`ARR_TIME`, `DEP_TIME`, `OP_CARRIER`, `TAIL_NUM`, `DISTANCE`, `DEP_DELAY`, `ARR_DELAY`) show extremely low cardinality ratios (<0.1%), reflecting repeated usage of common codes or repeated measurements across many flights. These fields are good candidates for categorical encoding in modeling.
3. Hourly and daily weather fields (`HourlyPrecipitation`, `HourlyDryBulbTemperature`, `DailyMaximumDryBulbTemperature`, and others) have very low cardinality ratios (<0.03%), indicating repeated measurements across flights sharing the same station and timestamp.
4. Columns like DATE (flight date) or `HourlySkyConditions` have moderate cardinality (~2-3%), representing recurring values over multiple flights or weather reports. These columns can be useful for time-series grouping or weather pattern analysis.
5. Highly unique columns have high cardinality ratios (6-20%), reflecting nearly unique values per flight. These are mostly identifiers, timestamps, or free-text fields.

#### Outlier Analysis

A comprehensive outlier analysis was performed on the OTPW dataset to quantify the prevalence of extreme values across flight and weather-related features. Consistent with prior EDA findings, weather features such as HourlyVisibility, HourlyWindGustSpeed, and HourlySeaLevelPressure show significant outlier percentages. This aligns with earlier observations that these features naturally exhibit extreme readings due to operational weather events rather than measurement errors. 

For flight features, temporal and scheduling variables show no outliers, consistent with expectations since they are bounded or categorical. In contrast, delay and operational metrics such as `DEP_DELAY_NEW`, `ARR_DELAY_NEW`, `DEP_DEL15`, and `ARR_DEL15` display the highest outlier rates. Other operational variables show small to moderate outlier presence reflecting variability in congestion, route characteristics, and airport-specific operational conditions.


![](https://raw.githubusercontent.com/Pinkaban/FlightDelayPrediction/main/Phase%20II%20Report%20Images/sthakur_phase2_eda_outliers_analysis.png)

The outlier distribution confirms that extreme values are concentrated in operationally meaningful features rather than in static identifiers or scheduled times. 
- For weather features, extreme values in visibility, wind, gusts, and precipitation reflect real events that are crucial predictors of flight delays, reinforcing the decision not to remove statistical outliers from these columns. 
- For flight performance metrics, outliers predominantly appear in delay and elapsed-time variables, consistent with prior findings that the delay distributions are inherently skewed with frequent extreme delays. Moderate outliers in taxi times and route distances indicate normal operational variability, while extreme delay components underscore occasional severe disruptions, including carrier, NAS, or late aircraft causes.

Overall, these outliers are not noise but represent meaningful operational extremes; the events that predictive models must capture to forecast delays accurately. Preprocessing strategies should thus retain these values, potentially applying transformations or robust modeling techniques to handle skewness without discarding the predictive signal embedded in extreme events.


#### Overall Learnings

Exploratory Data Analysis of the OTPW 12-month dataset, combined with weather and airport datasets, provided a comprehensive understanding of flight delays and their drivers. This analysis informed feature selection, highlighted patterns in the data, and motivated the creation of a custom joined dataset to better capture relevant interactions. Key learnings include:

1. **Flight delays are primarily carrier-driven:** Carrier-related issues account for nearly half of all delays, while weather contributes only about 3%, emphasizing that airline operational factors are dominant predictors.

2. **Temporal patterns are significant:** Delays exhibit strong seasonality and time-of-day effects. For instance, Q2 and June have above-average delays (+11% and +28% respectively), while SeptemberNovember are below average, highlighting the importance of including quarter, month, and time-based features.

![](https://raw.githubusercontent.com/Pinkaban/FlightDelayPrediction/main/Phase%20II%20Report%20Images/w261_project_phase2_otpw_temporal_patterns.png)

3. **Airline- and airport-specific effects:** Certain carriers (e.g., Frontier Airlines) and hub airports (e.g., ORD, ATL) experience delays substantially above the average, while others remain below average, suggesting operational idiosyncrasies are critical for predictive modeling.

![](https://raw.githubusercontent.com/Pinkaban/FlightDelayPrediction/main/Phase%20II%20Report%20Images/w261_project_phase2_otpw_airport_route.png)

4. **Route-level variability:** Low-volume routes show high variability in delays, whereas high-volume routes converge around the average (~20%), underscoring the need for route-level interaction features rather than simple volume proxies.

## Modeling

### Methodology and  Design


#### Modeling Pipeline and Loss Function
To handle the scale of the dataset, a distributed Logistic Regression architecture was implemented via PySpark. The pipeline minimized the **Binary Cross-Entropy (Log Loss)** function, augmented with an **L2 Regularization** (Ridge) penalty to ensure model stability and prevent overfitting to historical noise. 

The objective function is formally defined as:
$$J(w) = - \sum_{i=1}^{N} \left[ y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right] + \frac{\lambda}{2} \|w\|_2^2$$

Where \\( y_i\\) is the ground truth label, \\( \hat{y}_i\\) is the predicted probability of delay, and \\( \lambda\\) acts as the volume dial for the L2 regularization penalty, shrinking overly dominant feature coefficients.

#### Design
Our evaluation framework relied on strict temporal splits (Train: Jan-Sep, Validation: Oct, Test: Nov-Dec) to prevent data leakage. 



There were two primary design consideration to calibrate the model:

* **Concept Drift & Temporal Cross-Validation:** Standard randomized cross-validation ignores the chronological nature of weather and airline scheduling. We evaluated two temporal strategies: an *Expanding Window* (retaining all history) versus a *Sliding Window* (dropping older seasons as the window moves forward). The Sliding Window effectively mitigated concept drift. Concurrently, we tuned the regularization hyperparameter (`regParam`), selecting the model that maximized the Area Under the Precision-Recall Curve (PR-AUC) across the temporal folds.
* **Cost-Sensitive Threshold Optimization:** The default classification threshold of \\( 0.5\\) assumes equal business costs for false positives and false negatives. Because failing to predict an actual delay is highly expensive in aviation, we swept probability thresholds across the isolated October Validation set. The threshold was optimized to maximize the \\( F_2 Score \\), a metric that explicitly weights Recall twice as heavily as Precision.

##### Optimal Threshold selected: 0.30

#### Results
The empirically tuned threshold and optimal \\( \lambda\\) penalty were locked and applied to the final, untouched Test Set (November and December). 

By evaluating on the real-world distribution, the model successfully achieved a high **Recall**, catching the vast majority of actual delays. While this defensive posture naturally yielded a lower **Precision** (increasing false alarms), it perfectly aligned with the operational mandate mapped out by the \\( F_2 Score \\). Finally, we analyzed the learning curve efficiency—plotting training versus validation error across expanding data volumes—which confirmed that our undersampled model successfully converged without suffering from high variance or excessive bias.

### Tackling Class Imbalance: Dynamic Undersampling



Flight departure delays are an inherently skewed phenomenon, with a standard real-world imbalance of approximately 80% on-time to 20% delayed flights. Training a model natively on this distribution biases the algorithm toward majority-class prediction. 

To resolve this at scale, we implemented a memory-safe **Dynamic Random Undersampling** strategy, bypassing the computational overhead of dynamic class weighting. By randomly undersampling the majority class (on-time flights), we forced the training algorithm to learn from a perfectly balanced 1:1 environment. 
* *Critical Constraint:* Undersampling was applied **strictly to the training sets** (both globally and within specific cross-validation folds). Validation and Test partitions were explicitly shielded from this transformation, preserving their true, real-world distributions to ensure our evaluation metrics remained unbiased.

### Baseline Model

### Evaluation Metrics

With roughly 80% of flights departing on time, a naive model could achieve high accuracy by simply predicting "no delay" every time. Accuracy alone is therefore misleading for this problem. Instead, we adopted the F2 score as our primary metric, supplemented by F1 and PR-AUC for a fuller picture.

The F2 score is a variant of the F-beta family that computes the harmonic mean of precision and recall, but places twice as much weight on recall. This directly reflects our business priority: identifying delayed flights early enough for the operations team to act on gate planning, crew adjustments, and passenger communication. Missing a real delay is far more costly than issuing a false alarm, so we intentionally optimized for high recall. PR-AUC, or the area under the Precision-Recall curve, complements F2 by summarizing the model's ability to maintain precision across all recall levels, particularly useful when classes are imbalanced, since it is not inflated by the large number of correctly predicted on-time flights the way ROC-AUC can be.

Our final model captures 89% of actual delays on the test set. Precision is naturally lower as a result, but that tradeoff aligns with the operational mandate, it is safer to flag a potential disruption than to miss one entirely.

On the held-out test set (Nov–Dec), the model achieved a 0.55 F2 score and a 0.40 F1 score.


$$Precision = \frac{TP}{TP + FP}$$

$$Recall = \frac{TP}{TP + FN}$$

$$F_2 = \frac{5 \cdot Precision \cdot Recall}{4 \cdot Precision + Recall}$$

$$\text{PR-AUC} = \sum_{k=1}^{n} \left( R_k - R_{k-1} \right) \cdot P_k$$

Where:
* \\(TP\\) = True Positives (correctly predicted delays)
* \\(FP\\) = False Positives (false alarms)
* \\(FN\\) = False Negatives (missed delays)
* \\(P_k\\), \\(R_k\\) = Precision and Recall at the \\(k\\)-th decision threshold

* **The Massive Impact of Flight Feature Engineering:** The most significant performance jump in the entire pipeline occurred when transitioning from *Flights Baseline* (PR-AUC 0.2624) to *Flights Engineered* (PR-AUC 0.3539). This 35% relative improvement proves that engineered aviation features—such as historical route congestion, cascading delay metrics, and specific airline performance trends—are overwhelmingly the strongest drivers for predicting departure delays.
* **Weather is a Catalyst, Not a Sole Predictor:** Relying solely on weather features (whether baseline or engineered) yielded the poorest performance (PR-AUC ~0.23). On its own, the model struggles to differentiate standard weather from disruptive weather, resulting in a model that simply defaults to sounding the alarm for almost every flight (Recall 1.00, Precision 0.18). However, when engineered weather features are *combined* with engineered flight features (Experiment 6), they provide the necessary contextual boost to push the model to its peak performance (PR-AUC 0.3595).
* **Baseline Synergy vs. Engineered Synergy:** Simply combining the raw baseline flight and weather data (Experiment 4) yielded a PR-AUC of 0.2812. Our fully engineered combined dataset (Experiment 6) outperformed it significantly (PR-AUC 0.3595). This validates the entire Phase 1 and Phase 2 feature engineering lifecycle: raw data is insufficiently nuanced to model airline operations effectively.
* **Consistent F2-Score Optimization:** Across the highest-performing datasets, the threshold tuning strategy consistently protected the business mandate. In our final combined model, the pipeline secured an 89% delay catch rate while maintaining a strict 1:1 training environment via dynamic undersampling.


### Scalability & Feasibility

The pipeline was built from the ground up with scale in mind. All data processing, feature engineering, and model training run on PySpark across a 5-node cluster (1 driver + 4 workers, each with 8 cores and 32 GB RAM), which comfortably handled the 62 million weather observations and 5.8 million flight records in our 1 year dataset. Expensive operations like the custom weather join, which involves station-neighbor fallback, sliding-window accumulations, and airport-level congestion scoring, rely on broadcast joins, checkpointing, and strategic repartitioning to keep shuffle costs manageable. Hyperparameter tuning across 5 temporal folds completed in roughly 12 minutes, and threshold optimization took under 2 minutes, both well within practical limits for an iterative development cycle.

Scaling to the full 5 year dataset (2015–2019) in Phase III will multiply data volume by approximately 5×, but the architecture is designed for it. PySpark's distributed execution means we can scale horizontally by adding worker nodes rather than rewriting code. The memory-safe dynamic undersampling strategy avoids materializing the full majority class, so training costs grow linearly rather than quadratically with data size. The main bottleneck will be the custom weather join, which we plan to address by pre-computing station-level weather features once and caching them as intermediate parquet checkpoints. Moving to more complex models like Random Forest and MLP will increase training time, but the temporal cross-validation framework and feature pipeline remain unchanged, we simply swap out the estimator. Overall, the project is feasible within the existing Databricks infrastructure and does not require any fundamental architectural changes to handle the planned scale-up.


## Conclusion / Next Steps

In Phase II, we established a robust distributed Logistic Regression pipeline with strict temporal splitting, dynamic undersampling, sliding-window cross-validation, and threshold optimization to maximize the F2 Score. Through systematic feature engineering on both flight and weather data, we improved the baseline F2 from `0.5281` to `0.5500`, confirming that derived operational and meteorological features meaningfully enhance delay prediction. These results validate the pipeline architecture and set a strong foundation for more expressive modeling in Phase III.

For Phase III, we will migrate to a custom joined multi-year dataset (2015–2019), with 2019 reserved as the held-out test set, and refine the feature engineering pipeline to compute all imputation statistics exclusively from training data to eliminate any remaining leakage risk. On the feature side, we plan to derive additional variables from weather, flight, and station data—including weather–congestion interaction effects, time-synchronized destination weather impacts, and airport network delay propagation—while conducting ablation studies to ensure a minimal, high-impact feature set. On the modeling side, we will expand beyond Logistic Regression to Random Forest, Multilayer Perceptron, and Ensemble methods, aiming to capture the nonlinear delay dynamics that a linear model cannot fully exploit.

## Looking Ahead: Custom Data Join

To enhance predictive modeling potential beyond what the standard OTPW dataset provides, we developed a custom flight-level dataset by integrating flight operations, weather, station, and airport-level information. This integration enables the creation of richer features that capture operational complexity, network effects, and environmental influences—factors not fully represented in the original OTPW data.

The primary objective of this effort was to construct a high-quality, station-aligned, and time-synchronized dataset that consolidates multiple sources into a single, analytically robust resource. Specifically, the dataset combines:

- **Flight operations data:** departure and arrival times, aircraft identifiers, route information, and observed delays.
- **Weather information:** both primary station measurements and data from neighboring stations to account for spatial variability.
- **Time-series weather accumulations:** rolling aggregates over 3h, 6h, 12h, 24h, and daily windows to capture evolving meteorological impacts.
- **Airport and network congestion metrics:** departure and arrival volumes, historical delay patterns, and route-specific congestion.
- **Route delay history and tail-level turnaround metrics:** capturing cascading delays and aircraft-specific operational constraints.

By creating this enriched, integrated dataset, we provide a single source of truth for subsequent feature engineering and modeling efforts. It captures both network delay influences and cumulative weather effects, facilitating more accurate predictions of flight delays while preserving operational and environmental context.

This approach aligns directly with the prior motivation: integrating multiple data sources allows for richer, predictive features such as cumulative weather impacts, route-level interactions, and network effects—features that standard OTPW data alone cannot provide.

#### High-Level Approach
The custom dataset was constructed through a systematic, scalable, and station-aligned approach designed to preserve operational context and maximize predictive signal. Key considerations included:

- **Metric-level fallback:** For each weather metric, the primary station measurement was used first; if missing, the corresponding metric from a neighboring station was used to maintain continuity without introducing cross-metric bias.
- **Accurate time alignment:** Weather data was aligned at minute-level granularity, and sliding windows were constructed on the primary station timeline to capture evolving environmental conditions.
- **No station mixing:** Primary and neighbor weather sources were kept separate prior to fallback to prevent dilution of signal and maintain interpretability.
- **Data cleaning and quality assurance:** Stations and airports underwent validation for missing values, consistency checks, and geographic alignment. Outliers in operational and weather metrics were flagged but retained where operationally meaningful.
- **Airport graph and neighbor delay features:** A network of connected airports was constructed to compute route- and tail-level delay propagation, enabling neighbor-based congestion metrics and cascading delay indicators.
- **Weather Features Mapped to Flights based on 2-hour time window:** Each flight was mapped to a weather feature that has max timestamp before the cutoff timestamp that is 2 hour before the scheduled departure timestamp

To efficiently handle the scale of the OTPW dataset, processing was carefully partitioned:

- **Year- and month-level partitioning:** This allowed the pipeline to process smaller chunks of data, reducing infrastructure overhead and improving Spark job performance.
- **Incremental weather accumulation cycles:** Instead of joining all features at once, the pipeline first extracted weather_station and weather_timestamp for each flight, then accumulated weather metrics over the required windows (3h, 6h, 12h, 24h) separately, including neighbor weather where appropriate. This modular approach simplified computation and reduced memory usage.
- **Validation against OTPW:** The custom dataset was systematically compared to the original OTPW dataset to verify coverage, ensure no loss of flights, and confirm the integrity of key operational and weather features.

##### Step 1: Prepare Input Datasets

- NOAA ISD weather dataset (**df_weather**)

The weather dataset is built with data from NOAA having over 132 million observations at specific weather stations. While monthly and daily measurement fields are mostly missing, the high-frequency hourly weather readings allow us to align conditions around each flight's departure and arrival.

- FAA station-neighbor lookup (**df_stations**)

It includes over 5 million station-to-station neighbor relatinships. After removing the self-links, and including station - neighbor relationship for only the ones where the distance <=80, we select 34424 records.  
- FAA airport metadata (**df_airports**)

We use the Airport Codes dataset to map airports to geographic coordinates. Out of more than 9,000 records, we filter down to 1972 airports that are in US, are open and have valid IATA codes to support clean joins with the flights.
- BTS flight dataset (**df_flights**)

Flights dataset is built with the airline data from the DOT that has over 14 million flight records with rich operational details. Each flight is uniquely defined by a combination of date, carrier, flight number, and airport pair, although the raw data contains duplicate entries that we handle during preprocessing.

##### Step 2: Build Weather Features for Each Station


- **Base features with neighbor-station fallback**

To capture localized weather effects and improve feature completeness, weather data from neighboring stations is incorporated alongside the primary station. When a primary station measurement is missing for a given metric, corresponding values from nearby stations (within 80 km) are used to fill the gaps. This metric-level fallback ensures that each weather feature retains continuity without introducing cross-metric bias. By combining primary and neighbor station data, we generate a more robust and representative snapshot of the environmental conditions affecting each flight.
- **Sliding time-window weather accumulation**

Weather metrics are aggregated over rolling windows of 3h, 6h, 12h, and 24h for each station. These time-windowed features capture short- and medium-term weather trends, reflecting how changing conditions may influence flight operations. The accumulations include both primary and neighbor station data, allowing the model to consider both immediate and spatially adjacent meteorological impacts.
- **Daily weather feature aggregation**

For longer-term patterns, weather features are also accumulated on a daily basis for each station. This daily aggregation provides context for trends that may affect multiple flights, such as persistent visibility reductions, precipitation events, or temperature anomalies. This daily accumulation is different than the daily weather feature in OTPW in the way that this aggregates the feature from 12 am that day to 2 hours before the flight departure time.

Following weather features were selected for accumulation across time windows


| Column                  | Reason                                                          | Accum Method |
| ----------------------- | --------------------------------------------------------------- | ------------ |
| **HourlyPrecipitation** | Direct rainfall accumulation                                    | `sum`        |
| **HourlyWindGustSpeed**       | Extren wind affects safety. Use max of last hours | `max`        |
| **HourlyWindSpeed**       | Sustained wind matters for runway ops. Use average | `mean`        |
| **HourlyDryBulbTemperature**       | Temperature trend | `mean`        |
| **HourlyWetBulbTemperature**       | Correlates with icing and runway conditions | `mean`        |
| **HourlyDewPointTemperature**       | Fog and humidity | `mean`        |
| **HourlyRelativeHumidity**       | Fog, icing, and moisture | `mean`        |
| **HourlyVisibility**       | Low visibility drives delays | `min`        |



##### Step 3: Add Flight Features for each Flight

- **Airport-level delay features**

Delay patterns at the origin airport provide important context for understanding local operational disruptions. For each flight, we compute airport-level delay features over recent time windows (e.g., 1h, 3h, 6h), including average departure delay, maximum departure delay, and total number of departing flights. These features capture short-term congestion, ongoing operational issues, and delay clustering effects that often propagate across flights departing from the same airport.

- **Congestion and traffic volume features**

To represent dynamic airport congestion, departure and arrival volumes are aggregated over sliding windows surrounding each flight—typically from two hours prior to one hour after the scheduled departure time. These metrics reflect real-time airport load, enabling the model to distinguish between low-traffic and peak-traffic operational periods. Congestion signals often correlate strongly with taxi times, departure delays, and downstream disruptions.

- **Route history and delay propagation features**

Flights operating along the same route often experience correlated disruptions due to weather patterns, airspace constraints, or cascading delays. To capture this, we compute recent delay histories for the origin–destination pair, including maximum delays and delay averages over the past several hours. These features help the model identify route-specific vulnerabilities and patterns of repeat disruptions.

- **Tail-level turnaround features using airport graph**

Using the airport dataset, we construct an airport connectivity graph that maps aircraft identifiers (tails) across sequential flights. Tail-level turnaround metrics, such as previous flight delay, ground time between flights, and prior route performance—are computed to detect delay propagation across aircraft rotations. These features capture operational dependencies where a delayed inbound aircraft can directly influence the departure performance of the subsequent flight.

Following flight delay feature were created:

| Feature Type                  | Description                                                          | Features |
| ----------------------- | --------------------------------------------------------------- | ------------ |
| **Airport-level Delays**       | Rolling delay averages and number of flights in last hours | `dep_delay_1h`, `dep_delay_3h`, `dep_delay_6h`, `num_flights_3h`, `max_dep_delay_3h`        |
| **Neighboring Airport Delay** | Average delay and weighted delay by inverse distance                | `neighbor_delay_3h_avg`, `neighbor_weighted_delay`,`neighbor_flight_count`       |
| **Route-level Network Delays**       | Delays of other flights on same route (same origindest) in last 6h   | `route_avg_delay_6h`, `route_avg_delay_6h`        |
| **Network Congestion Delays**       | Rolling congestion based on departures and arrivals scheduled between past 2 hour to next one hour| `congestion_score`        |
| **Aircraft Turnaround Delays**       | Aircraft tail-number historic delays | `tail_avg_delay_10h`, `tail_recent_delay`        |

##### Step 4: Map Flights to Nearest ORIGIN and DEST stations

- **Map ORIGIN and DEST airports using IATA codes**

Each flight is first linked to its corresponding origin and destination airports by resolving the IATA airport codes to the standardized airport reference dataset. This mapping ensures consistent geographic attributes such as latitude, longitude, and elevation and provides a clean foundation for integrating weather and operational features.

- **Identify nearest weather stations for each airport**

Using the geospatial metadata (latitude and longitude) from both airports and NOAA weather stations, we compute the nearest weather stations for each airport based on great-circle distance. This station-airport mapping is essential for accurately associating weather conditions with flight operations. The nearest station is designated as the primary station, while additional nearby stations—typically within an 80 km radius—are tagged as neighbor stations to provide fallback and spatial context.

- **Assign each flight to its closest relevant weather station**

For both the origin and destination of every flight, the pipeline maps the airport to its nearest weather station(s). These station IDs become the anchors for aligning each flight with temporally matched weather observations. This mapping ensures that environmental features (base readings, accumulations, and neighbor fallbacks) are accurately tied to the meteorological conditions experienced at the corresponding airport at the time of departure or arrival.

##### Step 5: Integrate Flights with Weather Data

- **Determine the cutoff time for each flight**

For every flight, a weather cutoff timestamp is computed based on the scheduled departure time. This ensures that only weather conditions known prior to departure are used in modeling, preserving temporal integrity and avoiding any form of information leakage. The cutoff logic typically subtracts a small operational buffer (e.g., minutes prior to scheduled departure) to reflect the real-world availability of weather observations.

- **Select the most recent valid weather observation within the cutoff window**

Using the mapped primary (and fallback neighbor) weather stations, the pipeline retrieves all weather records up to the cutoff timestamp and selects the latest available measurement. This produces a precise, time-aligned snapshot of flight-specific weather conditions—reflecting what would have been operationally observable at the moment of decision-making.

- **Extract weather features for the aligned timestamp**

Once the latest weather timestamp is identified, all associated weather features—base metrics, sliding-window accumulations, and daily aggregates—are joined to the flight record. This ensures that each flight is enriched with a consistent and complete set of weather attributes, incorporating both immediate and cumulative meteorological effects.

- **Generate the final flight–weather integrated dataset**

The resulting dataset, containing flight operations, airport/station mappings, and fully aligned weather features, is saved as the final integrated flights-with-weather table. This dataset forms the foundation for downstream exploratory analysis, feature engineering, and predictive modeling, serving as a single, high-quality source for all subsequent analytical workflows.

### Custom Feature Engineering

Building on the enriched flight–weather dataset, several advanced feature engineering strategies were applied to capture complex interactions and temporal effects that influence flight delays. These custom features leverage the combination of operational data, airport network structures, and time-synchronized weather information to provide deeper predictive signals that go beyond standard delay metrics.

- **Weather-Airport Congestion Interaction Effects:** Using flight data, airport graphs, and weather severity metrics, these features quantify how adverse weather interacts with airport congestion to amplify delays, capturing nonlinear effects that can significantly enhance model performance.
- **Time-Synchronized Weather Impact:** By aligning flight timestamps with both origin and destination weather data, these features account for scenarios where departure weather is favorable but arrival conditions are severe, capturing potential ATC or en-route delay contributions.
- **Weather Lead/Lag Effects:** Leveraging station-wise rolling weather accumulations (3h, 6h, 12h), these features measure the influence of recent past weather on current flight performance, enabling the model to detect delayed effects of evolving weather patterns.
- **Airport Network Delay Propagation:** By constructing airport connectivity graphs and linking sequential flights, these features quantify how delays in inbound flights propagate to subsequent departures, modeling operational contagion effects.
- **Multi-Modal Delay Drivers:** Combining network-based and weather-based features, these metrics extract higher-order patterns, capturing complex interactions between environmental and operational factors that drive delays.

These custom features enrich the dataset with temporal, network, and interaction-based insights, enabling the model to account for cascading effects, localized weather impacts, and nonlinear dependencies. Collectively, they provide a strong foundation for improved delay prediction and operational analysis.
